```python
import pandas as pd
from datetime import datetime

# Path to your CSV file
input_file = 'transaction_data.csv'  # Replace with your actual file path
output_file = 'transaction_data_converted.csv'

# Read the CSV file
df = pd.read_csv(input_file)
```

```python
df.head()
```

```python
# Specify the name of the date column (change 'TRANSACTION_DATE' to your actual column name)
date_column = 'TransactionTime'  # Update with your actual date column name

# Function to convert date to ISO 8601 format
def convert_to_iso(date_str):
    try:
        # Parse the date using the specific format
        parsed_date = datetime.strptime(date_str, '%a %b %d %H:%M:%S IST %Y')
        return parsed_date.strftime('%Y-%m-%d %H:%M:%S')
    except ValueError:
        return None  # Return None for invalid dates

# Apply the function to the date column
df[date_column] = df[date_column].apply(convert_to_iso)

# Save the updated DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print(f"Date conversion completed. Saved to {output_file}")
```
